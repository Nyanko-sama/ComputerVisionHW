### Step 0

Run the necessary imports.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Step 1

Load an image (any image that contains faces).

```python
image = cv2.imread("data/R.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### Step 2

Load the dlib face predictor.

```python
import dlib

# Let's load the detector
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
faces = detector(gray)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
print('Number of detected faces:', len(faces))
import random

# Draw rectangle around each face
def plot_result(image, faces):
    result_dlib = np.copy(image)
    for rect in faces:    
        # Draw rectangle around the face
        x_1, y_1, x_2, y_2 = (rect.left(), rect.top(), rect.right(), rect.bottom())
        color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
        cv2.rectangle(result_dlib, (x_1, y_1), (x_2, y_2), color, 3)
    plt.imshow(result_dlib)

plot_result(image, faces)
```

### Step 5 (optional)

Repeat the process with a different and more challenging image (more faces, smaller faces, people with glasses, hats, helmets, etc.). How does the detector perform? Is it robust?

```python
pop = cv2.imread("data/pop.jpeg")
gray_pop = cv2.cvtColor(pop, cv2.COLOR_BGR2GRAY)
pop = cv2.cvtColor(pop, cv2.COLOR_BGR2RGB)

plot_result(pop, detector(gray_pop))

# Detects everybody in the image, and also a person with glasses.
```

```python
we = cv2.imread("data/we.jpg")
gray_we = cv2.cvtColor(we, cv2.COLOR_BGR2GRAY)
we = cv2.cvtColor(we, cv2.COLOR_BGR2RGB)

plot_result(we, detector(gray_we))

# Detects even cartoon characters. 
```
